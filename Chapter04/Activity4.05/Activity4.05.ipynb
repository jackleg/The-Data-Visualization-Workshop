## Activity 4.05: Linear Regression for Animal Attribute Relations
You have a dataset about various animals including their body mass and maximum longevity. To discover if there is any linear relationship between these two variables a regression plot shall be used.

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mydata = pd.read_csv("../../Datasets/anage_data.csv")
```

```python
mydata.head()
```

Filter the data so you end up with samples containing a body mass and a maximum longevity. Only consider samples for the class Mammalia and a body mass below 200,000.

```python
longevity = 'Maximum longevity (yrs)'
mass = 'Body mass (g)'
data = mydata[mydata['Class'] == 'Mammalia']
data = data[np.isfinite(data[longevity]) & np.isfinite(data[mass]) & (data[mass] < 200000)]
```

```python
data
```

Create a regression plot to visualize the linear relationship of the variables.

```python
sns.set()
plt.figure(figsize=(10, 6), dpi=300)
sns.regplot(mass, longevity, data=data)
```

```python
# Create figure
sns.set()
plt.figure(figsize=(10, 6), dpi=300)
# Create scatter plot
sns.regplot(mass, longevity, data=data)
# Show plot
plt.show()
```
